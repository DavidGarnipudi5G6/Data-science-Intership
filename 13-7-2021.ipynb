### join/merge
* inner,left.right,outer

```python
import pandas as pd
import numpy as np
d={'sno':[101,102,103,120,121],
  'sname':['a','b','c','h','i']
  }
df1=pd.DataFrame(d)
df1
```

```python
d2={'sno':[101,102,103,120,121,115,161],
  'sname':['aa','bb','cc','ee','i','gg','kk']
  }
df2=pd.DataFrame(d2)
df2
```

```python

m=pd.merge(df1,df2,on='sno')
m
```

```python
m=pd.merge(df1,df2,on='sno',how='left')
m
```

```python
m=pd.merge(df1,df2,on='sno',how='right')
m
```

```python
m=pd.merge(df1,df2,on='sno',how='outer')
m
```

### missing values

* fillna
* dropna
* replace
* SimpeImputer

```python
a=np.array([[1,2,np.nan,3,4],
          [10,12,13,14,15],
          [34,67,89,90,12],
          [np.nan,10,23,np.nan,34],
          [89,np.nan,31,56,np.nan],
          [90,np.nan,np.nan,78,89]])
a
```

```python
df=pd.DataFrame(a,columns=['one','two','three','four','five'])
df
```

```python
# find the null values in dataset

df.isnull().sum().sum()
```

```python
# print the null values rows from dataset
is_nan = df.isnull()
df
```

```python
# remove the all null values from dataset using dropna
# dropna : it is useful when large dataset
# dropna : it is loss when low data set beacuse you loss the important data
df.dropna()
```

```python
# replace
df['one']=df['one'].replace(np.nan,0)
df
```

```python
df['two'].mean()
```

```python
df['two']=df['two'].fillna(df['two'].mean())
df
```

```python
df['three'].median()
```

```python
df['three']=df['three'].fillna(df['three'].median())
df
```

```python
# ffill => forword fill
df.fillna(method='ffill')
```

```python
# bfill => backward fill
df.fillna(method='bfill')
```

```python
a=np.array([[1,2,np.nan,3,4],
          [10,12,13,14,15],
          [34,67,89,90,12],
          [np.nan,10,23,np.nan,34],
          [89,np.nan,31,56,np.nan],
          [90,np.nan,np.nan,78,89]])
df=pd.DataFrame(a,columns=['one','two','three','four','five'])
df
```

```python
# fill the missing values using scikit-learn(simple Imputer)
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='mean')
d=s.fit_transform(df)
df=pd.DataFrame(d)
df
```

```python
from sklearn.impute import SimpleImputer
s=SimpleImputer(strategy='median')
d=s.fit_transform(df)
df=pd.DataFrame(d)
df
```
